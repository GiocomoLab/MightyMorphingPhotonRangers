```python
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
import numpy as np
import scipy as sp
from scipy.ndimage.filters import gaussian_filter1d
import sqlite3 as sql
import os
import pandas as pd
from datetime import datetime
from glob import glob
os.sys.path.append('../')
from utilities import *
from astropy.convolution import convolve, Gaussian1DKernel

%load_ext autoreload
%autoreload 2

%matplotlib inline
```

```python
# find sessions from single mouse on TwoTower_noTimeout with datestring earlier than or equal to first timeout day
# num rewards greater than 
df = load_session_db()
mouse = '4054011.2'
# mouse filter
df = df[df['MouseName'].str.match(mouse)]
df = df[df['Track']=='TwoTower_Timeout']
df = df[df['DateFolder']=='17_07_2018']
df
```

```python
data_pre = behavior_dataframe([df['data file'].iloc[1], df['data file'].iloc[2]],concat=True)
```

```python
trial_mat, bin_edges, bin_centers = make_pos_bin_trial_matrices(data_pre[['speed','morph','lick rate','reward','lick']]._values,
                                              data_pre['pos']._values,
                                              data_pre['tstart']._values,
                                              data_pre['teleport']._values,bin_size=5)


trial_mat[np.isnan(trial_mat)]=0
morph_vec,count = sp.stats.mode(trial_mat[:,:,1],axis=1,nan_policy='omit')
morph_vec = np.squeeze(morph_vec)

#print(trial_mat[:,:-1])
hits = np.squeeze(np.nanmax(trial_mat[:,:,-1],axis=1))
hits[hits>0]=1
```

```python

speed_dict = trial_type_dict(trial_mat[:,:,0],morph_vec)
f,ax = plot_speed(bin_centers,speed_dict,morph_vec)
f.savefig('muscimolFigs/speed_control.svg',format='svg')
f.savefig('muscimolFigs/speed_control.pdf',format='pdf')
```

```python
lick_dict = trial_type_dict(trial_mat[:,:,4],morph_vec)
lick_mat = np.squeeze(trial_mat[:,:,2])
lick_mat_norm = lick_dict['all']/np.amax(lick_dict['all'])
lick_norm_dict = trial_type_dict(lick_mat_norm,morph_vec)
fig,axes = lick_plot(lick_norm_dict,bin_edges,smooth=False,ratio=False)
fig.savefig('muscimolFigs/lick_raster_control.svg',format='svg')
fig.savefig('muscimolFigs/lick_raster_control.pdf',format='pdf')
```

```python
trial_info_pre = by_trial_info(data_pre)
pcnt_mean_pre = avg_by_morph(trial_info_pre['morphs'],trial_info_pre['pcnt'])
```

```python
trial_info_post = by_trial_info(data_post)
pcnt_mean_post = avg_by_morph(trial_info_post['morphs'],trial_info_post['pcnt'])
```

```python
f,ax = plt.subplots(figsize=[5,5])
morph_vals = np.arange(0,1.25,.25)
ax.plot(morph_vals,pcnt_mean_pre,color='black')
ax.plot(morph_vals,pcnt_mean_post,color='red')
ax.set_ylabel("P(licked at second tower)")
ax.set_xlabel("morph")
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.legend(('control','muscimol'),loc='lower right')
f.savefig('muscimolFigs/percent_correct.svg',format='svg')
f.savefig('muscimolFigs/percent_correct.pdf',format='pdf')
```

```python
zone0_speed_pre, zone0_speed_post = avg_by_morph(trial_info_pre['morphs'],trial_info_pre['zone0_speed']), avg_by_morph(trial_info_post['morphs'],trial_info_post['zone0_speed'])
zone1_speed_pre, zone1_speed_post = avg_by_morph(trial_info_pre['morphs'],trial_info_pre['zone1_speed']), avg_by_morph(trial_info_post['morphs'],trial_info_post['zone1_speed'])

f,ax = plt.subplots(1,2,figsize=[10,5])
morph_vals = np.arange(0,1.25,.25)
ax[0].plot(morph_vals,zone0_speed_pre,color='black')
ax[0].scatter(trial_info_pre['morphs'],trial_info_pre['zone0_speed'],color='black',s=5)
ax[0].plot(morph_vals,zone0_speed_post,color='red')
ax[0].scatter(trial_info_post['morphs']+.03,trial_info_post['zone0_speed'],color='red',s=5)
ax[0].set_ylabel("cm/s")
ax[0].set_xlabel("morph")
ax[0].set_title("zone 0 speed")
ax[0].spines['top'].set_visible(False)
ax[0].spines['right'].set_visible(False)


ax[1].plot(morph_vals,zone1_speed_pre,color='black')
ax[1].scatter(trial_info_pre['morphs'],trial_info_pre['zone1_speed'],color='black',s=5)
ax[1].plot(morph_vals,zone1_speed_post,color='red')
ax[1].scatter(trial_info_post['morphs']+.03,trial_info_post['zone1_speed'],color='red',s=5)
ax[1].set_ylabel("cm/s")
ax[1].set_xlabel("morph")
ax[1].set_title("zone 1 speed")
ax[1].spines['top'].set_visible(False)
ax[1].spines['right'].set_visible(False)

f.savefig('muscimolFigs/speed_by_morph.svg',format='svg')
f.savefig('muscimolFigs/speed_by_morph.pdf',format='pdf')
```
