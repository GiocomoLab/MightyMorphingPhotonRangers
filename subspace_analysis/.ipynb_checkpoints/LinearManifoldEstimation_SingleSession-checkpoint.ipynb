```python

def load_ca_mat(fname):
    ca_dat = {}

    try:
        with h5py.File(fname,'r') as f:
            for k,v in f.items():       
                try:
                    ca_dat[k] = np.array(v)
                except:
                    print(key + "not made into numpy array")
                    ca_dat[k]=v

    except:    
        ca_dat = sp.io.loadmat(CaDataDir+ CaFile+"_cnmf_results.mat")
        for key in ca_dat.keys():
            if isinstance(ca_dat[key],np.ndarray):
                ca_dat[key] = ca_dat[key].T
    return ca_dat
```

```python
fname= CaDataDir+ CaFiles[i] +"_cnmf_results.mat"
print(fname)
ca_dat = load_ca_mat(fname)
```

# PCA - use all timepoints 

```python
pca = PCA()
C_dec = sp.stats.zscore(ca_dat['C_dec'][behavior['ca_inds'].values,:],axis=0)
C_pca = pca.fit_transform(C_dec)
```

```python
# skree plot
f,axarr = plt.subplots(2,3,figsize=[30,15])

axarr[0,0].plot(pca.explained_variance_)
axarr[0,0].set_ylabel("% variance",size=28)
axarr[0,0].tick_params(labelsize=20)

axarr[0,1].plot(pca.singular_values_)
axarr[0,1].set_ylabel("eigenvalue",size=28)
axarr[0,1].tick_params(labelsize=20)

axarr[0,2].plot(np.log(pca.singular_values_))
axarr[0,2].set_ylabel("log(eigenvalue)",size=28)
axarr[0,2].tick_params(labelsize=20)


axarr[1,0].plot(pca.explained_variance_[:10])
axarr[1,0].set_ylabel("% variance",size=28)
axarr[1,0].tick_params(labelsize=20)

axarr[1,1].plot(pca.singular_values_[:10])
axarr[1,1].set_ylabel("eigenvalue",size=28)
axarr[1,1].tick_params(labelsize=20)

axarr[1,2].plot(np.log(pca.singular_values_[:10]))
axarr[1,2].set_ylabel("log(eigenvalue)",size=28)
axarr[1,2].tick_params(labelsize=20)
```

```python
trialMask = behavior['position'].values>0
context_trialMask = (behavior['position'].values>0) & (behavior['morph'].values!=0) &(behavior['morph'].values!=1)

f,axarr = plt.subplots(2,2,figsize=[15, 15])

s_cxt=axarr[0,0].scatter(C_pca[trialMask,1],C_pca[trialMask,2],c=behavior['morph'].values[trialMask],cmap='cool',s=2)
plt.colorbar(s_cxt,ax=axarr[0,0])
axarr[0,0].set_title('Colored by Context')
axarr[0,0].set_xlabel("PC 1")
axarr[0,0].set_ylabel("PC 2")

s_pos=axarr[0,1].scatter(C_pca[trialMask,1],C_pca[trialMask,2],c=behavior['position'].values[trialMask],cmap='magma',s=2)
plt.colorbar(s_pos,ax=axarr[0,1])
axarr[0,1].set_title('Colored by Position')
axarr[0,1].set_xlabel("PC 1")
axarr[0,1].set_ylabel("PC 2")

s_cxt=axarr[1,0].scatter(C_pca[trialMask,2],C_pca[trialMask,3],c=behavior['morph'].values[trialMask],cmap='cool',s=2)
axarr[1,0].set_xlabel("PC 2")
axarr[1,0].set_ylabel("PC 3")

s_cxt=axarr[1,1].scatter(C_pca[trialMask,2],C_pca[trialMask,3],c=behavior['position'].values[trialMask],cmap='magma',s=2)
axarr[1,1].set_xlabel("PC 2")
axarr[1,1].set_ylabel("PC 3")
```
