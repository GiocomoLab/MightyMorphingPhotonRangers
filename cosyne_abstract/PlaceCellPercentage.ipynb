```python
f,ax = plt.subplots(2,3,figsize=[10,6])
M = 0
for mouse in tt_mice:
    ax[0,0].plot(frac[mouse][0],color=plt.cm.cool(0),alpha=.4)
    ax[0,0].plot(frac[mouse][1],color=plt.cm.cool(1.),alpha=.4)
    ax[0,0].set_title('Timeout Mice')
    
    ax[1,0].plot(np.array(frac[mouse][0])-np.array(frac[mouse][1]))
    ax[0,0].set_ylabel('Place Cell Fraction')
    ax[1,0].set_ylabel("Difference")
for mouse in reversal_mice:
    ax[0,1].plot(frac[mouse][0],color=plt.cm.cool(0),alpha=.4)
    ax[0,1].plot(frac[mouse][1],color=plt.cm.cool(1.),alpha=.4)
    ax[0,1].set_title('Reversal Mice')
    ax[1,1].plot(np.array(frac[mouse][0])-np.array(frac[mouse][1]))
    
for mouse in notask_mice:
    ax[0,2].plot(frac[mouse][0],color=plt.cm.cool(0),alpha=.4)
    ax[0,2].plot(frac[mouse][1],color=plt.cm.cool(1.),alpha=.4)
    ax[0,2].set_title('No Timeout Mice')
    
    ax[1,2].plot(np.array(frac[mouse][0])-np.array(frac[mouse][1]))
    
for i in range(ax.shape[1]):
    ax[0,i].set_ylim([0,1])
    ax[0,i].set_xlim([0,10])
    ax[1,i].set_ylim([-.3,.3])
    ax[1,i].set_xlim([0,10])

    
f.savefig("G:\\My Drive\\Figures\\TwoTower\\COSYNE2019\\PlaceCellFrac.pdf",format='pdf')
```
